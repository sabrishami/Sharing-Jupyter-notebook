```python
from sklearn import linear_model
```

```python
lr=linear_model.LinearRegression()
X_train=np.array(train_data['sqft_living'], dtype=pd.Series).reshape(-1,1)
```

```python
Y_train=np.array(train_data['price'],dtype=pd.Series)
lr.fit(X_train,Y_train)
```

```python
X_test=np.array (test_data['sqft_living'], dtype=pd.Series).reshape(-1,1)
Y_test=np.array(test_data['price'], dtype=pd.Series)
pred=lr.predict(X_test)
```

```python

cv=format(cross_val_score(lr,data[['sqft_living']], data['price'], cv=5).mean(),'.3f')
rmsesm = float(format(np.sqrt(metrics.mean_squared_error(Y_test,pred)),'.3f'))
rtrsm= float(format(lr.score(X_train,Y_train),".3f"))
rtesm=float(format(lr.score(X_test,Y_test),'.3f'))
```

```python
print("CrossValidation =", cv)
```

```python
print('Intercept:', format(lr.intercept_))
#print('Coefficient:' format(lr.coef_))
```

```python
df_dm=data
```

```python
#Binning
#take the year from the date column
df_dm['sales_yr']=df_dm['date'].astype(str).str[:4]
# add the age of bulidings as new column when the houses were sold 
df_dm['age']= df_dm['sales_yr'].astype(int)-df_dm['yr_built']
df_dm['age'][df_dm['age']<0]= 0

    

```

```python
# add the age of the renovation as new column when the houses were sold 

df_dm['age_rnv']=0
df_dm['age_rnv']=df_dm['sales_yr'][df_dm['yr_renovated']!=0].astype(int)-df_dm['yr_renovated'][df_dm['yr_renovated']!=0].astype(int)
df_dm['age_rnv'][df_dm['age_rnv'].isnull()]=0
```

```python
df_dm
```

```python
df_dm["age"].max()
df_dm["age"].min()
# partioning the age into bins
bins=[-1,0,5,10,25,50,75,100,100000]
labels=["<1","1-5",'6-10','11-25','26-50','51-75','76-100','>100']
df_dm['age-binned']=pd.cut(df_dm['age'],bins=bins,labels=labels)

# partition the age_rnv into bins
bins=[-1,0,5,10,25,50,75,100000]
labels=["<1","1-5",'6-10','11-25','26-50','51-75','>76']
df_dm['age_rnv-binned']=pd.cut(df_dm['age_rnv'],bins=bins,labels=labels)

# histograms for the binned columns

f, axes=plt.subplots(1,2,figsize=(15,5))
p1=sns.countplot(df_dm['age-binned'], ax=axes[0])
for p in p1.patches:
    height = p.get_height()
    p1.text(p.get_x()+p.get_width()/2,height + 50,height,ha="center") 

    

p2=sns.countplot(df_dm['age_rnv-binned'],ax=axes[1])
sns.despine(left=True, bottom=True)
for p in p2.patches:
    height = p.get_height()
    p2.text(p.get_x()+p.get_width()/2,height + 200,height,ha="center")
    

axes[0].set(xlabel='Age')
axes[0].yaxis.tick_left()
axes[1].yaxis.set_label_position("right")
axes[1].yaxis.tick_right()
axes[1].set(xlabel='Renovation Age');

df_dm = pd.get_dummies(df_dm, columns=['age-binned','age_rnv-binned'])
```
